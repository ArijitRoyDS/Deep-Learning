# Convert Color Scales

```python
# Read an image in RGB
image2 = cv2.imread("Data\opencv.jpg", 1)
image2
```

```python
# Convert BGR to RGB

rgb_image = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
cv2.imshow("Girl Image", rgb_image )
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Convert BGR to Luv

rgb_image1 = cv2.cvtColor(image2, cv2.COLOR_BGR2Luv)
cv2.imshow("Girl Image", rgb_image1)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Convert BGR to Grayscale

rgb_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
cv2.imshow("Girl Image", rgb_image2)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Convert BGR to HSV

rgb_image3 = cv2.cvtColor(image2, cv2.COLOR_BGR2HSV)
cv2.imshow("Girl Image", rgb_image3)
cv2.waitKey()
cv2.destroyAllWindows()
```

# Define a function to perform the above operations

```python
def viewImage(image, name_of_window):
    cv2.namedWindow(name_of_window, cv2.WINDOW_NORMAL)
    cv2.imshow(name_of_window, image)
    cv2.waitKey()
    cv2.destroyAllWindows()
```

```python
img = cv2.imread('Data\Modi.jpg', 1)
viewImage(img, "Modi Ji")
```

# Resize the Image

```python
image = cv2.imread('Data\opencv.jpg', 1)
image.shape
```

```python
# Resize based on fixed number of pixels

resized_image = cv2.resize(image, (1920,1080))
cv2.imshow("resized girl image", resized_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Resize by a specific percentage

scale_percent = 30
width = int(image.shape[1]*scale_percent/100)
height = int(image.shape[0]*scale_percent/100)
dim = (width, height)
resized_image = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
viewImage(resized_image, "After resizing the image to 30%")
```

```python
# Resize to half of original size

resized_image2 = cv2.resize(image, (int(image.shape[1]/2),int(image.shape[0]/2)))
cv2.imshow("resized image",resized_image2)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Resize to 75% of original size:

resized_image3 = cv2.resize(image, (int(image.shape[1]*0.75),int(image.shape[0]*0.75)))
cv2.imshow("resized image",resized_image3)
cv2.waitKey()
cv2.destroyAllWindows()
```

# Image Rotation 

```python
image = cv2.imread('Data\opencv.jpg', 1)
viewImage(image,  "Girl")
```

```python
# Rotate by 45 Degrees from Center
(h, w, d) = image.shape
center = (w//2, h//2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(image, M, (w,h))
viewImage(rotated,  "image after rotation")
```

```python
# Rotate by 180 Degrees from Center
(h, w, d) = image.shape
center = (w//2, h//2)
M = cv2.getRotationMatrix2D(center, 180, 1.0)
rotated = cv2.warpAffine(image, M, (w,h))
viewImage(rotated,  "image after rotation")
```

```python
# Rotate by 145 Degrees from 1st Quadrant
(h, w, d) = image.shape
center = (w//4, h//4)
M = cv2.getRotationMatrix2D(center, 145, 1.0)
rotated = cv2.warpAffine(image, M, (w,h))
viewImage(rotated,  "image after rotation")
```
